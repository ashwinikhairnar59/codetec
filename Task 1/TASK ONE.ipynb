```python
import pandas as pd

# Load the dataset
file_path = "dataset1.csv"
data = pd.read_csv(file_path)

# Display the first few rows
print("Dataset Overview:")
print(data.head())
```

```python
# Display basic information
print("\nDataset Information:")
print(data.info())

# Summary statistics
print("\nSummary Statistics:")
print(data.describe())
```

```python
# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical columns
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[col], kde=True, bins=20, color='blue')
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Bar plots for categorical columns
categorical_cols = data.select_dtypes(include=['object']).columns

for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    data[col].value_counts().plot(kind='bar', color='orange')
    plt.title(f"Value Counts of {col}")
    plt.show()
```

```python
# Scatter plots for numerical features against the target
target_col = "Target"  # Target column in the dataset

for col in numerical_cols:
    if col != target_col:
        plt.figure(figsize=(8, 4))
        sns.scatterplot(x=data[col], y=data[target_col], color='green')
        plt.title(f"Scatter Plot: {col} vs {target_col}")
        plt.show()
```

```python
# Correlation matrix for numerical features
corr_matrix = data[numerical_cols].corr()

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Boxplots for detecting outliers in numerical columns
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data[col], color='red')
    plt.title(f"Boxplot of {col}")
    plt.show()
```
